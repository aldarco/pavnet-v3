## baseband method

```python
data = pd.read_csv("/data/PAVNET_26/PLO-v3//ANTAR_amplitude_2026-01-08-to-2026-01-08.csv", 
                   parse_dates=True, index_col=0)
```

```python
data.index += timedelta(hours=5)
data.head()
```

```python
# traken from pavnet.c / it wa not yet implemented
# SCALE_FACTOR = ( (1000.0f * RPgainValue_cfg) / (2147483648.0f * calib_voltage));
SCALE_FACTOR = ( (1000.0 * 1) / (2147483648.0 * 1));
data_db = 20*np.log10(data * SCALE_FACTOR)
```

```python
fig, ax = plt.subplots()
data_db.NAA.loc[dt(2026, 1, 18):dt(2026, 1, 19)].plot(ax=ax)
```
